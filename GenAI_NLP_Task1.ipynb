#Task 1: Conceptual Understanding

### 1. Difference between "Love" and "love" in NLP
In Natural Language Processing, text is case-sensitive. This means "Love" and "love" are treated as two different tokens. This increases the vocabulary size unnecessarily and can affect model performance. Therefore, text is usually converted to lowercase during preprocessing.

---

### 2. What happens if stopwords are not removed?
If stopwords are not removed:
- The dataset contains unnecessary common words like "is", "the", "and"
- It increases computation time
- It may reduce model accuracy by adding noise

---

### 3. Scenarios where removing stopwords can be harmful
1. **Sentiment Analysis**  
   Example: "I am not happy" → removing "not" changes the meaning.

2. **Question Answering Systems**  
   Example: "What is AI?" → removing "what" removes the intent.

---

### 4. Difference between Stemming and Lemmatization

- **Stemming**: Cuts words to base form using simple rules  
  Example: "running" → "run"

- **Lemmatization**: Converts words to meaningful base form using vocabulary  
  Example: "better" → "good"

- Stemming is faster but less accurate  
- Lemmatization is slower but more accurate

#Task 2:Build Advanced Preprocessing Function

```python
import re

def preprocess_text(text):

    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    text = text.lower()

    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)

    # Fix repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    # Final correction for small words
    tokens = [re.sub(r'(.)\1{1,}', r'\1', w) if len(w) <= 3 else w for w in tokens]

    tokens = [w for w in tokens if (len(w) > 2 or w in ['no', 'not'])]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

```python
sample = "WOW!!! I have 2 dogs!!! Visit http://example.com soooo goooood!!! 😍"

tokens, clean = preprocess_text(sample)

print("Tokens:", tokens)
print("Cleaned Sentence:", clean)
```

```python
sample = "soooo goooood!!!"
print(preprocess_text(sample))
```

#Task 3: Stress Testing

```python
test_data = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
for text in test_data:

    result = preprocess_text(text)   # safe call

    tokens = result[0]
    clean = result[1]

    print("Original Text:", text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean)
    print("-" * 50)
```

#Task 4: Token Analytics

```python
def token_analysis(tokens):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens == 0:
        avg_length = 0
    else:
        avg_length = sum(len(word) for word in tokens) / total_tokens

    return total_tokens, unique_tokens, avg_length
```

```python
for text in test_data:

    result = preprocess_text(text)
    tokens = result[0]

    total, unique, avg = token_analysis(tokens)

    print("Original Text:", text)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg, 2))
    print("-" * 50)
```

#Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []

for text in test_data:

    result = preprocess_text(text)
    tokens = result[0]

    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 most frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[-5:])
```

#Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:

        result = preprocess_text(text)
        tokens = result[0]
        clean = result[1]

        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
output = full_pipeline(test_data)

print("All Tokens:")
print(output["tokens"])

print("\nClean Sentences:")
for sentence in output["clean_sentences"]:
    print(sentence)
```
